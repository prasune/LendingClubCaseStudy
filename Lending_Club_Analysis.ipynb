### Binning: convert continuous variable to categorical varaible

```python
#low-income: 0-50000
#mid-income: 50001-100000
#high-income: 100001>
def annul_inc_conversion(x):
  if x<= 50000:
    return "low income"
  elif x> 50000 and x <= 100000:
    return "medium income"
  else:
    return "high income"
```

```python
loan_df["annual_inc_range"] = loan_df["annual_inc"].apply(lambda x:annul_inc_conversion(x))
loan_df.head()
```

```python
loan_df["int_rate"].describe()
```

```python
loan_df["int_rate"]
```

```python
#low-interest: 7.5
#mid-interest: 7.6-13
#high-interest: 13+
def interest_rate_conversion(x):
  if x<= 7.5:
    return "low interest rate"
  elif x>7.5  and x <= 13.0:
    return "medium interest rate"
  else:
    return "high interest rate"
```

```python
loan_df["int_rate_range"]= loan_df["int_rate"].apply(lambda x:interest_rate_conversion(x) )
loan_df.head()
```

```python
# Remove columns not relevant for the analysis
# Remove emp_title, title, revol_bal,revol_util
unnecessary_columns = ["emp_title", "title", "revol_bal", "revol_util"]
loan_df.drop(unnecessary_columns,axis=1,inplace=True)
loan_df.shape
```

```python
# Update tenure of loan in number of years
loan_df['term'] = loan_df['term'].apply(lambda x: int(x.replace(' months', ''))/12)
loan_df['term']
```

```python
loan_df['loan_amnt'].describe()
```

```python
# loan_amnt < 5000 - low
# loan_amnt < 10000 - medium
# loan_amnt < 15000 - high
# loan_amnt > 15000 - very high
def loan_amount_range(loan_amount):
    if loan_amount < 5000:
        return "low"
    elif loan_amount < 10000:
        return "medium"
    elif loan_amount < 15000:
        return "high"
    else:
        return "very high"
```

```python
loan_df['loan_amnt_range'] = loan_df['loan_amnt'].apply(lambda x: loan_amount_range(x))
loan_df['loan_amnt_range']
```

```python
loan_df['emp_length'].describe()
```

```python
# emp_length == 0 - just employed
# emp_length <=2 - employed
# emp_length <= 4 - stable
# emp_length <= 9 - good stability
# emp_length > 9 - very good stability
def emp_length_range(emp_length):
    if emp_length == 0:
        return "just employed"
    elif emp_length <= 2:
        return "employed"
    elif emp_length <= 5:
        return "stable"
    elif emp_length < 10:
        return "good stability"
    else:
        return "very good stability"
```

```python
loan_df['emp_length_range'] = loan_df['emp_length'].apply(lambda x: emp_length_range(x))
loan_df['emp_length_range']
```
